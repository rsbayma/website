### *Alias*, pré-filtros antialias

**Teorema da amostragem**: qual deve ser a taxa de amostragem?

Sinal contínuo: $x(t)$


Trem de impulsos:
$$
    p(t) = \sum_{k=-\infty}^\infty \delta(t-kT)
$$

Série de Fourier:
$$
    p(t) = \frac{1}{T}\sum_{k=-\infty}^\infty e^{jn\omega_0t}
$$
onde $\omega_0 = \frac{2\pi}{T}$

Sinal amostrado:
$$
\begin{align}
    \hat{x}(t) &= x(t)p(t) \\
    &= \frac{1}{T}\sum_{n=-\infty}^\infty x(t)\,e^{jn\omega_0 t}
\end{align}
$$

Espectro do sinal amostrado:
$$
\begin{align}
    \hat{X}(\omega) &= \frac{1}{T}\sum_{n=-\infty}^\infty \mathcal{F}\left\{x(t)\,e^{jn\omega_0 t}\right\}\\
    &=\frac{1}{T}\sum_{n=-\infty}^\infty X(\omega - n\omega_0)
\end{align}
$$

Interpretação:

```python
import numpy as np
from matplotlib import pyplot as plt

X = lambda w: (1-w**2)*(np.abs(w)<=1).astype(np.float32)
ww = np.linspace(-10,10,500)
N = 3
Xh = np.zeros(ww.shape)
w0 = 2
for n in range(N+1):
    if n==0:
        Xh = X(ww)
    else:
        Xh += X(ww-n*w0)+X(ww+n*w0)

plt.figure(figsize=(15,8))
plt.subplot(211)
plt.plot(ww,X(ww))
plt.subplot(212)
plt.plot(ww,Xh)
plt.xticks(np.arange(-10,11,2))
plt.show()
```
